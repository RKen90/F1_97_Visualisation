## Amount of retirements across the season

```python
# Number of full-time teams and races
full_time_teams = 10
full_time_races = 17
cars_per_team = 2

# Calculate the number of competing cars for full-time teams
full_time_entries = full_time_teams * cars_per_team * full_time_races

# Part-time team entries
part_time_team_entries = 2  # 2 cars
part_time_races = 2  # 2 weekends

# Calculate the total entries for the part-time team
part_time_entries = part_time_team_entries * part_time_races

# Total number of competing cars across the season
total_competing_cars = full_time_entries + part_time_entries

print(f"Total number of competing cars across the season: {total_competing_cars}")
```

```python
# Assuming your transposed DataFrame is `df_transposed`

# Select only the rows containing race results (excluding any summary rows like 'oints')
race_results = drivers_transposed.iloc[:-1, :]

# Flatten the race results into a single list (this turns the DataFrame into a single column of values)
flattened_results = race_results.values.flatten()

# Convert to a pandas Series to easily apply value_counts
flattened_series = pd.Series(flattened_results)

# Get the value counts for each outcome (1st, 2nd, 3rd, Ret, etc.)
result_counts = flattened_series.value_counts()

# Display the counts for 1st, 2nd, 3rd, and Ret
print(result_counts[['Ret']])
```

```python
# Replace "Ret" with 0 and ensure the race result columns are converted to numeric
drivers_97_cleaned = drivers_97.replace('Ret', 0)

# Convert columns (race results) to numeric
drivers_97_cleaned.iloc[:, 2:-1] = drivers_97_cleaned.iloc[:, 2:-1].apply(pd.to_numeric, errors='coerce')

# Check the cleaned data
drivers_97_cleaned.head()
```

```python
# Define the points mapping
points_map = {1: 10, 2: 6, 3: 4, 4: 3, 5: 2, 6: 1}

# Apply the points mapping
drivers_97_points = drivers_97_cleaned.iloc[:, 2:-1].applymap(lambda x: points_map.get(x, 0) if pd.notnull(x) else 0)

# Display the mapped points
drivers_97_points.head()

```

```python
# Calculate the cumulative points race-by-race
cumulative_points = drivers_97_points.cumsum(axis=1)

# Display the cumulative points
cumulative_points

```

```python
# Step 1: Extract the driver names from your original dataframe
drivers = drivers_97['Driver']

# Step 2: Set the driver names as the index for the cumulative points dataframe
cumulative_points.index = drivers

cumulative_points_cleaned = cumulative_points.drop('Driver', axis = 0)

# Step 3: Verify that the driver names are now the index
cumulative_points_cleaned
```
